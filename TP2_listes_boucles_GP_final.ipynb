## 1.1 Les Listes en Python

### Qu'est-ce qu'une liste ?

Une **liste** est une collection ordonn√©e de donn√©es. En g√©nie des proc√©d√©s, on les utilise pour stocker :
- Des mesures exp√©rimentales (temp√©rature, pression, d√©bit)
- Des compositions de m√©langes
- Des r√©sultats de calculs

### Cr√©ation de listes

````python
# Temp√©ratures mesur√©es dans un r√©acteur (¬∞C)
temperatures = [25, 50, 75, 100, 125]
print("Temp√©ratures :", temperatures)

# Pressions mesur√©es (bar)
pressions = [1.0, 1.5, 2.0, 2.5, 3.0]
print("Pressions :", pressions)

# Compositions molaires
compositions = [0.25, 0.35, 0.20, 0.15, 0.05]
print("Compositions :", compositions)
````

### Acc√©der aux √©l√©ments d'une liste

Les indices commencent √† **0** en Python !

````python
mesures = [25.5, 26.1, 25.8, 26.3, 25.9]

# Premier √©l√©ment (indice 0)
print("Premi√®re mesure :", mesures[0])

# Deuxi√®me √©l√©ment (indice 1)
print("Deuxi√®me mesure :", mesures[1])

# Dernier √©l√©ment (indice -1)
print("Derni√®re mesure :", mesures[-1])

# Avant-dernier √©l√©ment (indice -2)
print("Avant-derni√®re mesure :", mesures[-2])
````

### Op√©rations de base sur les listes

````python
mesures = [25.5, 26.1, 25.8, 26.3, 25.9]

# Longueur de la liste (nombre d'√©l√©ments)
print("Nombre de mesures :", len(mesures))

# Somme des √©l√©ments
print("Somme :", sum(mesures))

# Maximum
print("Maximum :", max(mesures))

# Minimum
print("Minimum :", min(mesures))

# Moyenne (somme / nombre)
moyenne = sum(mesures) / len(mesures)
print(f"Moyenne : {moyenne:.2f}")
````

### Modifier une liste

````python
temperatures = [25, 50, 75]
print("Liste initiale :", temperatures)

# Ajouter un √©l√©ment √† la fin
temperatures.append(100)
print("Apr√®s append(100) :", temperatures)

# Ins√©rer un √©l√©ment √† la position 1
temperatures.insert(1, 37.5)
print("Apr√®s insert(1, 37.5) :", temperatures)

# Supprimer un √©l√©ment par valeur
temperatures.remove(37.5)
print("Apr√®s remove(37.5) :", temperatures)

# Supprimer le dernier √©l√©ment
temperatures.pop()
print("Apr√®s pop() :", temperatures)

# Trier la liste
temperatures.sort()
print("Apr√®s sort() :", temperatures)
````

## 1.2 Le Slicing (D√©coupage)

Le **slicing** permet d'extraire une partie d'une liste.

**Syntaxe** : `liste[debut:fin:pas]`

- `debut` : indice de d√©part (inclus)
- `fin` : indice de fin (exclu)
- `pas` : intervalle entre les √©l√©ments (optionnel)

````python
temperatures = [20, 30, 40, 50, 60, 70, 80, 90, 100]
print("Liste compl√®te :", temperatures)

# Les 3 premiers √©l√©ments
print("3 premiers :", temperatures[:3])

# Du 3√®me au 6√®me √©l√©ment
print("Indices 2 √† 5 :", temperatures[2:6])

# Les 3 derniers √©l√©ments
print("3 derniers :", temperatures[-3:])

# Un √©l√©ment sur deux
print("Un sur deux :", temperatures[::2])

# Liste invers√©e
print("Invers√©e :", temperatures[::-1])
````

## 1.3 Les Boucles `for`

La boucle `for` permet de **parcourir** une liste et d'effectuer des op√©rations sur chaque √©l√©ment.

### Syntaxe de base

````python
temperatures = [25, 50, 75, 100]

# Parcourir la liste
for T in temperatures:
    print(f"Temp√©rature : {T}¬∞C")
````

### Utilisation de `range()`

`range(n)` g√©n√®re une s√©quence de 0 √† n-1

````python
# Afficher les nombres de 0 √† 4
for i in range(5):
    print(i)

print()

# De 1 √† 10
for i in range(1, 11):
    print(i, end=' ')

print("\n")

# De 0 √† 10 par pas de 2
for i in range(0, 11, 2):
    print(i, end=' ')
````

### Utilisation de `enumerate()`

`enumerate()` donne √† la fois l'**indice** et la **valeur**

````python
temperatures = [25, 50, 75, 100]

# Avec enumerate (d√©marre √† 0)
for index, T in enumerate(temperatures):
    print(f"Mesure {index} : {T}¬∞C")

print()

# Commencer √† 1
for index, T in enumerate(temperatures, 1):
    print(f"Mesure {index} : {T}¬∞C")
````

### Utilisation de `zip()`

`zip()` permet de parcourir **plusieurs listes en m√™me temps**

````python
temperatures = [25, 50, 75, 100]
pressions = [1.0, 1.5, 2.0, 2.5]

# Parcourir les deux listes ensemble
for T, P in zip(temperatures, pressions):
    print(f"T = {T}¬∞C, P = {P} bar")
````

## 1.4 Les Boucles `while`

La boucle `while` continue tant qu'une **condition** est vraie.

**‚ö†Ô∏è Attention** : Il faut s'assurer que la condition devient fausse, sinon boucle infinie !

````python
# Exemple : Mont√©e en pression
pression = 1.0  # bar
temps = 0       # minutes

print("Mont√©e en pression :")
while pression < 5.0:
    print(f"t = {temps} min, P = {pression} bar")
    pression += 0.5  # Augmentation de 0.5 bar
    temps += 1       # Chaque minute

print(f"Pression cible atteinte : {pression} bar")
````

## 1.5 Application : Conversion d'unit√©s

Exemple pratique : Convertir des temp√©ratures de ¬∞C en Kelvin

````python
# Temp√©ratures en Celsius
T_celsius = [25, 50, 75, 100, 125]

# M√©thode 1 : Avec une boucle for
T_kelvin = []
for T in T_celsius:
    T_K = T + 273.15
    T_kelvin.append(T_K)

print("Conversions ¬∞C ‚Üí K :")
for celsius, kelvin in zip(T_celsius, T_kelvin):
    print(f"{celsius}¬∞C = {kelvin} K")

print()

# M√©thode 2 : List comprehension (plus concis)
T_kelvin_v2 = [T + 273.15 for T in T_celsius]
print("M√©thode 2 :", T_kelvin_v2)
````
