### 1. Qu'est-ce qu'une variable ?

Une **variable** est comme une bo√Æte qui stocke une valeur. En Python, on cr√©e une variable en lui donnant un nom et en lui assignant une valeur avec le signe `=`.

**R√®gles pour nommer une variable :**
- Commence par une lettre (a-z, A-Z) ou un underscore (_)
- Peut contenir des lettres, chiffres et underscores
- Pas d'espaces (utilisez _ √† la place)
- Sensible √† la casse : `Temperature` ‚â† `temperature`
- √âvitez les mots r√©serv√©s de Python (if, for, while, etc.)

````python
# Exemples de d√©claration de variables
temperature = 25.5
pression = 101325
nom_reacteur = "R√©acteur A"
est_actif = True

print("Temp√©rature :", temperature)
print("Pression :", pression)
````

### 2. Les types de donn√©es fondamentaux

Python utilise principalement 4 types de donn√©es de base :

| Type | Description | Exemple |
|------|-------------|----------|
| `int` | Nombre entier | `42`, `-15`, `1000` |
| `float` | Nombre d√©cimal | `3.14`, `-0.5`, `2.0` |
| `str` | Cha√Æne de caract√®res | `"Bonjour"`, `'Python'` |
| `bool` | Bool√©en (Vrai/Faux) | `True`, `False` |

Pour conna√Ætre le type d'une variable, on utilise la fonction `type()`.

````python
# Exemples de types
nombre_entier = 10
nombre_decimal = 3.14
texte = "G√©nie des proc√©d√©s"
vrai_ou_faux = True

print("Type de nombre_entier:", type(nombre_entier))
print("Type de nombre_decimal:", type(nombre_decimal))
print("Type de texte:", type(texte))
print("Type de vrai_ou_faux:", type(vrai_ou_faux))
````

### 3. Op√©rations arithm√©tiques

Python permet d'effectuer toutes les op√©rations math√©matiques classiques :

| Op√©ration | Symbole | Exemple | R√©sultat |
|-----------|---------|---------|----------|
| Addition | `+` | `5 + 3` | `8` |
| Soustraction | `-` | `5 - 3` | `2` |
| Multiplication | `*` | `5 * 3` | `15` |
| Division | `/` | `5 / 2` | `2.5` |
| Division enti√®re | `//` | `5 // 2` | `2` |
| Modulo (reste) | `%` | `5 % 2` | `1` |
| Puissance | `**` | `5 ** 2` | `25` |

````python
# Exemples d'op√©rations
a = 10
b = 3

print("Addition:", a + b)
print("Soustraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Division enti√®re:", a // b)
print("Modulo:", a % b)
print("Puissance:", a ** b)
````

### 4. Op√©rations sur les cha√Ænes de caract√®res

Les cha√Ænes de caract√®res (strings) se d√©finissent avec des guillemets simples `'...'` ou doubles `"..."`.

**Op√©rations possibles :**
- **Concat√©nation** : Assembler deux cha√Ænes avec `+`
- **R√©p√©tition** : R√©p√©ter une cha√Æne avec `*`
- **Longueur** : Obtenir le nombre de caract√®res avec `len()`

````python
# Exemples avec les cha√Ænes
prenom = "Marie"
nom = "Curie"

# Concat√©nation
nom_complet = prenom + " " + nom
print("Nom complet:", nom_complet)

# R√©p√©tition
print("R√©p√©tition:", "Python " * 3)

# Longueur
print("Longueur du nom:", len(nom_complet))
````

### 5. Conversion de types (Casting)

Parfois, il faut convertir une variable d'un type √† un autre. Python propose des fonctions de conversion :

- `int()` : Convertit en entier
- `float()` : Convertit en d√©cimal
- `str()` : Convertit en cha√Æne de caract√®res
- `bool()` : Convertit en bool√©en

````python
# Exemples de conversion
nombre_texte = "42"
nombre_entier = int(nombre_texte)
print("Converti en entier:", nombre_entier, "Type:", type(nombre_entier))

decimal = float("3.14")
print("Converti en float:", decimal, "Type:", type(decimal))

entier = 100
texte = str(entier)
print("Converti en string:", texte, "Type:", type(texte))
````

L'√©nergie n√©cessaire se calcule avec : **Q = m √ó c √ó ŒîT**

````python
# creation des variables :
Ti=20
Tf=85
m=50
c=4.18
# calcul de dT :
dT=Tf-Ti
# conversion de la masse en grammes :
m_g=m*1000
# calcul de l'√©nergie n√©cessaire :
Q=m_g*c*dT
# convesion en kJ :
Q_kJ=Q/1000
# affichage de resultats :
print(f"dT= {dT}¬∞C")
print(f"masse= {m_g}g")
print(f"energie en J= {Q}J")
print(f"energie en kJ= {Q_kJ}kJ")
````
