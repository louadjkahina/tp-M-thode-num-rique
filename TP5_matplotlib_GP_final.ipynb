## 1.1 Introduction √† Matplotlib

### Pourquoi Matplotlib en G√©nie des Proc√©d√©s ?

- ‚úÖ **Visualiser des donn√©es exp√©rimentales**
- ‚úÖ **Analyser des tendances**

### Import

````python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

print("Pr√™t pour la visualisation !")
````

## 1.2 Premier Graphique - Courbe Simple

### √âvolution de temp√©rature dans un r√©acteur

````python
# Donn√©es
temps = np.array([0, 10, 20, 30, 40, 50, 60])  # minutes
temperature = np.array([25, 35, 48, 65, 82, 95, 105])  # ¬∞C

# Cr√©er le graphique
plt.figure(figsize=(10, 6))
plt.plot(temps, temperature)
plt.xlabel('Temps (min)')
plt.ylabel('Temp√©rature (¬∞C)')
plt.title('√âvolution de la temp√©rature dans le r√©acteur')
plt.grid(True, alpha=0.3)
plt.show()
````

## 1.3 Personnalisation

### Couleurs, marqueurs et styles

````python
# Donn√©es
temps = np.array([0, 10, 20, 30, 40, 50, 60])
temperature = np.array([25, 35, 48, 65, 82, 95, 105])

# Graphique personnalis√©
plt.figure(figsize=(10, 6))
plt.plot(temps, temperature,
         color='red',           # Couleur
         marker='o',            # Marqueurs circulaires
         linestyle='--',        # Ligne pointill√©e
         linewidth=2,           # √âpaisseur de ligne
         markersize=8,          # Taille des marqueurs
         label='Temp√©rature')   # L√©gende

plt.xlabel('Temps (min)', fontsize=12)
plt.ylabel('Temp√©rature (¬∞C)', fontsize=12)
plt.title('Mont√©e en temp√©rature - R√©acteur batch', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## 1.4 Plusieurs Courbes sur un M√™me Graphique

### Temp√©rature et Pression

````python
# Donn√©es
temps = np.array([0, 10, 20, 30, 40, 50, 60])
temperature = np.array([25, 35, 48, 65, 82, 95, 105])
pression = np.array([1.0, 1.2, 1.5, 1.8, 2.0, 2.1, 2.2])

# Graphique avec deux axes Y
fig, ax1 = plt.subplots(figsize=(12, 6))

# Premier axe Y : Temp√©rature
color = 'tab:red'
ax1.set_xlabel('Temps (min)', fontsize=12)
ax1.set_ylabel('Temp√©rature (¬∞C)', color=color, fontsize=12)
ax1.plot(temps, temperature, color=color, marker='o', linewidth=2, label='Temp√©rature')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, alpha=0.3)

# Deuxi√®me axe Y : Pression
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Pression (bar)', color=color, fontsize=12)
ax2.plot(temps, pression, color=color, marker='s', linewidth=2, label='Pression')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('√âvolution T et P dans le r√©acteur', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
````

## 1.5 Nuage de Points (Scatter)

### Corr√©lation entre variables

````python
# Donn√©es : Temp√©rature vs Conversion
temperature = np.array([50, 60, 70, 80, 90, 100])
conversion = np.array([15, 28, 45, 65, 80, 90])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(temperature, conversion, s=100, c='purple', alpha=0.6, edgecolors='black')
plt.xlabel('Temp√©rature (¬∞C)', fontsize=12)
plt.ylabel('Conversion (%)', fontsize=12)
plt.title('Effet de la temp√©rature sur la conversion', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
````

## 1.6 Diagrammes en Barres

### Comparaison de rendements

````python
# Donn√©es
conditions = ['Exp 1', 'Exp 2', 'Exp 3', 'Exp 4', 'Exp 5']
rendements = [75, 82, 68, 90, 85]

# Diagramme en barres
plt.figure(figsize=(10, 6))
plt.bar(conditions, rendements, color='steelblue', edgecolor='black')
plt.xlabel('Conditions exp√©rimentales', fontsize=12)
plt.ylabel('Rendement (%)', fontsize=12)
plt.title('Comparaison des rendements', fontsize=14)
plt.ylim(0, 100)
plt.axhline(y=80, color='r', linestyle='--', label='Objectif 80%')
plt.legend()
plt.grid(True, axis='y', alpha=0.3)
plt.show()
````

## 1.7 Histogrammes

### Distribution de mesures

````python
# G√©n√©ration de donn√©es (distribution normale)
np.random.seed(42)
mesures = np.random.normal(100, 5, 100)  # Moyenne 100, √©cart-type 5, 100 mesures

# Histogramme
plt.figure(figsize=(10, 6))
plt.hist(mesures, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Valeur mesur√©e', fontsize=12)
plt.ylabel('Fr√©quence', fontsize=12)
plt.title('Distribution des mesures', fontsize=14)
plt.axvline(x=np.mean(mesures), color='red', linestyle='--', linewidth=2, label=f'Moyenne = {np.mean(mesures):.1f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## 1.8 Subplots - Plusieurs Graphiques

### Dashboard de proc√©d√©

````python
# Donn√©es
temps = np.linspace(0, 60, 100)
T = 25 + 80 * (1 - np.exp(-temps/20))
P = 1.0 + 1.2 * (1 - np.exp(-temps/25))
X = 90 * (1 - np.exp(-temps/30))

# Cr√©er une grille 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Graphique 1 : Temp√©rature
axes[0, 0].plot(temps, T, 'r-', linewidth=2)
axes[0, 0].set_title('Temp√©rature', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Temps (min)')
axes[0, 0].set_ylabel('T (¬∞C)')
axes[0, 0].grid(True, alpha=0.3)

# Graphique 2 : Pression
axes[0, 1].plot(temps, P, 'b-', linewidth=2)
axes[0, 1].set_title('Pression', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Temps (min)')
axes[0, 1].set_ylabel('P (bar)')
axes[0, 1].grid(True, alpha=0.3)

# Graphique 3 : Conversion
axes[1, 0].plot(temps, X, 'g-', linewidth=2)
axes[1, 0].set_title('Conversion', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Temps (min)')
axes[1, 0].set_ylabel('X (%)')
axes[1, 0].axhline(y=90, color='r', linestyle='--', alpha=0.5)
axes[1, 0].grid(True, alpha=0.3)

# Graphique 4 : T vs X
axes[1, 1].scatter(T, X, c=temps, cmap='viridis', s=50, alpha=0.6)
axes[1, 1].set_title('T vs Conversion', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('T (¬∞C)')
axes[1, 1].set_ylabel('X (%)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
