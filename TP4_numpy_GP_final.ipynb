## 1.1 Introduction √† NumPy

### Pourquoi NumPy en G√©nie des Proc√©d√©s ?

- ‚úÖ **Rapidit√©** : 10-100x plus rapide que les listes Python
- ‚úÖ **Calculs vectoriels** : Op√©rations sur des tableaux entiers
- ‚úÖ **Alg√®bre lin√©aire** : R√©solution de syst√®mes, matrices
- ‚úÖ **Fonctions math√©matiques** : exp, log, sin, cos optimis√©es

### Import

````python
import numpy as np

print(f"Version NumPy : {np.__version__}")
````

## 1.2 Cr√©er des Arrays

### √Ä partir d'une liste

````python
# Temp√©ratures mesur√©es (¬∞C)
T_list = [25, 50, 75, 100, 125]
T_array = np.array(T_list)

print("Liste Python :", T_list)
print("Array NumPy :", T_array)
print("Type :", type(T_array))
````

### Arrays sp√©ciaux

````python
# Array de z√©ros (10 √©l√©ments)
zeros = np.zeros(10)
print("Z√©ros :", zeros)

# Array de uns (5 √©l√©ments)
ones = np.ones(5)
print("Uns :", ones)

# Array avec une valeur constante
constante = np.full(8, 3.14)
print("Constante :", constante)
````

### S√©quences avec arange et linspace

````python
# arange : de 0 √† 10 par pas de 2
T1 = np.arange(0, 11, 2)
print("arange(0, 11, 2) :", T1)

# linspace : 5 valeurs entre 0 et 100
T2 = np.linspace(0, 100, 5)
print("linspace(0, 100, 5) :", T2)

# Temp√©ratures de 20 √† 100¬∞C (11 points)
T3 = np.linspace(20, 100, 11)
print("Temp√©ratures :", T3)
````

## 1.3 Op√©rations Vectorielles

### Conversion Celsius ‚Üí Kelvin (vectoris√©e)

````python
# Temp√©ratures en Celsius
T_C = np.array([25, 50, 75, 100])

# Conversion en Kelvin (une seule op√©ration !)
T_K = T_C + 273.15

print("Celsius :", T_C)
print("Kelvin :", T_K)
````

### Op√©rations arithm√©tiques

````python
# D√©bits volumiques (m¬≥/h)
Q = np.array([1.0, 1.5, 2.0, 2.5])

# Masse volumique (kg/m¬≥)
rho = 1000

# D√©bits massiques (op√©ration vectorielle)
m_dot = rho * Q

print("D√©bits volumiques (m¬≥/h) :", Q)
print("D√©bits massiques (kg/h) :", m_dot)
````

### Fonctions math√©matiques

````python
# Concentrations (mol/L)
C = np.array([1.0, 4.0, 9.0, 16.0, 25.0])

# Racine carr√©e (vectoris√©e)
C_sqrt = np.sqrt(C)
print("‚àöC :", C_sqrt)

# Exponentielle
k = 0.1  # constante de vitesse
t = np.array([0, 10, 20, 30, 40])  # temps (min)
C_t = np.exp(-k * t)
print("\nD√©croissance exponentielle :")
print("t :", t)
print("C(t) :", C_t)
````

## 1.4 Statistiques

### Fonctions statistiques essentielles

````python
# Temp√©ratures mesur√©es
temperatures = np.array([25.2, 25.8, 24.9, 25.5, 25.3, 25.7, 25.1])

print("Temp√©ratures :", temperatures)
print(f"Moyenne : {np.mean(temperatures):.2f}¬∞C")
print(f"√âcart-type : {np.std(temperatures):.3f}¬∞C")
print(f"Minimum : {np.min(temperatures)}¬∞C")
print(f"Maximum : {np.max(temperatures)}¬∞C")
print(f"M√©diane : {np.median(temperatures)}¬∞C")
````

## 1.5 Indexation et Slicing

````python
temps = np.array([0, 10, 20, 30, 40, 50, 60])
T = np.array([25, 35, 48, 65, 82, 95, 105])

print("Donn√©es compl√®tes :")
print("t :", temps)
print("T :", T)
print()

# Phase de chauffe (3 premi√®res mesures)
print("Phase de chauffe :")
print("t :", temps[:3])
print("T :", T[:3])
print()

# Une mesure sur deux
print("Une mesure sur deux :")
print("t :", temps[::2])
print("T :", T[::2])
````

## 1.6 Matrices (Arrays 2D)

### Cr√©er une matrice

````python
# Matrice de coefficients st≈ìchiom√©triques
# R√©action : N2 + 3H2 ‚Üí 2NH3
stoechio = np.array([[1, 3, -2],   # N2, H2, NH3
                     [0, 1, 0]])    # Autre contrainte

print("Matrice st≈ìchiom√©trique :")
print(stoechio)
print(f"Forme : {stoechio.shape}")  # (2 lignes, 3 colonnes)
````

### Matrice identit√©

````python
# Matrice identit√© 3x3
I = np.eye(3)
print("Matrice identit√© :")
print(I)
````

### Produit matriciel

````python
A = np.array([[2, 1],
              [1, 3]])

B = np.array([[1],
              [2]])

# Produit matriciel
C = A @ B  # ou np.dot(A, B)

print("A =")
print(A)
print("\nB =")
print(B)
print("\nA @ B =")
print(C)
````

## 1.7 R√©solution de Syst√®mes Lin√©aires

### Exemple : Bilan de mati√®re

Syst√®me : $A \vec{x} = \vec{b}$

Solution : $\vec{x} = A^{-1} \vec{b}$

````python
# Syst√®me d'√©quations:
# 2x + y = 5
# x + 3y = 8

A = np.array([[2, 1],
              [1, 3]])

b = np.array([5, 8])

# R√©solution
x = np.linalg.solve(A, b)

print("Solution :")
print(f"x = {x[0]:.2f}")
print(f"y = {x[1]:.2f}")

# V√©rification
verification = A @ x
print(f"\nV√©rification : A @ x = {verification}")
print(f"Devrait √™tre : b = {b}")
````
